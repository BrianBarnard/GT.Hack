<h4>Loading the transactions:</h4>

```python
#Load the transactions 
df_transactions = pd.read_csv('Data/10000_transactions.csv',encoding="ISO-8859-1")

#Print the first few transactions
df_transactions.head()
```

<h4>Grouping the transactions by customer_id and order_number:</h4>

```python
#Group the transactions by order number 
df_group = df_transactions.set_index(['customer_id','order_number']).groupby(['order_number','customer_id'])

#Viewing the groupby data
df_group.head()
```

<h4>Finding the number of customers:</h4>

```python
#Number of customers 
len(set(df_transactions['customer_id']))
```

<h4>Finding the most popular items in the transactions:</h4>

```python
#Most popular transactions 
df_transactions['product_name'].value_counts().head()
```

<h4>Finding the most popular department id in the transactions:</h4>

```python
#Most popular department 
df_transactions['department_id'].value_counts().head()
```

<h4>Find the number of transactions that have purchased a particular item:</h4>

```python
#Number of transactions that have purchased Gluten Free Whole Grain Bread
item_name = 'Gluten Free Whole Grain Bread'
df_transactions[df_transactions['product_name'] == item_name].shape[0]
```
